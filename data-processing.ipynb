```python
import re
import pandas as pd
```

```python
f=open('_chatGroup.txt','r',encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
pattern = r'\[\d{2}/\d{2}/\d{2},\s\d{1,2}:\d{2}:\d{2}\s*(?:AM|PM)?\]\s*'
```

```python
messages= re.split(pattern,data)
len(messages)
```

```python
messages
```

```python
messages = [msg.replace("\u200e", "").replace("\n", " ").replace("\u202f","") for msg in messages]
```

```python
messages=messages[1:]
```

```python
messages
```

```python
dates= re.findall(pattern,data)
```

```python
dates=[date.replace("\u202f"," ") for date in dates]
```

```python
# dates=dates[1:]
```

```python
print(len(messages), len(dates))
```

```python
dates
```

```python
cleaned_dates = [date.strip("[] ") for date in dates]
df = pd.DataFrame({"User_message": messages, "Message_date": cleaned_dates})
df["Message_date"] = pd.to_datetime(df["Message_date"], format='%d/%m/%y, %I:%M:%S %p')
df.head()
```

```python
df.shape
```

```python
messages
```

```python
users=[]
messages=[]
for message in df["User_message"]:
    entry=re.split(r'([\w\W]+?):\s',message)
    if(entry[1:]):
        users.append(entry[1].strip())
        messages.append(entry[2].strip())
    else:
        users.append(r'Group notification')
        messages.append(entry[0].strip())

df["User"]=users
df["Message"]=messages
```

```python
df.drop(columns=["User_message"], inplace=True)
df.head()
```

```python
df['Year']=df['Message_date'].dt.year
```

```python
df.head()
```

```python
df['Month']=df['Message_date'].dt.month
df['Day']=df['Message_date'].dt.day
df['Hour']=df['Message_date'].dt.hour
df['Minute']=df['Message_date'].dt.minute
```
